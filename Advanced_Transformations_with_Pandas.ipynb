First we clean two csv files and combine them into one final dataFrame with columns:
- Hero (Just the name of the Hero)
- Publisher
- Gender
- Eye color
- Race
- Hair color
- Height (numeric)
- Skin color
- Alignment
- Weight (numeric)
- Plus, one-hot-encoded columns for every power that appears in the dataset. E.g.:
 - Agility
 - Flight
 - Superspeed
 - etc.

```python
## Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json
```

```python
df1 = pd.read_csv('superhero_info.csv')
df1.head()
```

```python
df1.info()
```

```python
df2 = pd.read_csv('superhero_powers.csv')
df2.head()
```

```python
df2.info()
```

```python
## save the Hero|Publisher column into 2 new columns into the dataframe
df1[['Hero','Publisher']] = df1['Hero|Publisher'].str.split('|',expand=True)
df1.head()
```

```python
## drop the original Hero|Publisher column 
df1 = df1.drop(columns=['Hero|Publisher'])
df1.head()
```

```python
## examining a single value from the Measurements column
Measurements = df1.loc[0,"Measurements"]
print(type(Measurements))
Measurements
```

```python
Measurements = Measurements.replace("'",'"')
Measurements
```

```python
## now we can use json.loads
fixed_Measurements = json.loads(Measurements)
print(type(fixed_Measurements))
fixed_Measurements
```

```python
## use .str.replace to replace all single quotes
df1['Measurements'] = df1['Measurements'].str.replace("'",'"')
## Apply the json.loads to the full column
df1['Measurements'] = df1['Measurements'].apply(json.loads)
df1['Measurements'].head()
```

```python
Height_Weight = df1['Measurements'].apply(pd.Series)
Height_Weight
```

```python
# concat Height_Weight with original dataframe
df1 = pd.concat((df1, Height_Weight), axis = 1)
df1.head()
```

```python
df1 = df1.drop(columns=['Measurements'])
df1
```

```python
df1['Height'] = df1['Height'].str.replace(' cm', '')
df1['Height'] = df1['Height'].astype(float)
```

```python
df1['Weight'] = df1['Weight'].str.replace(' kg', '')
df1['Weight'] = df1['Weight'].astype(float)
```

```python
df1.info()
```

```python
df1
```

```python
df2
```

```python
# Create a new column where the powers are split into a list
df2['Powers_split'] = df2['Powers'].str.split(',')
df2['Powers_split']
```

```python
## exploding the Powers column of lists
exploded = df2.explode('Powers_split')
exploded[['hero_names','Powers','Powers_split']].head(5)
```

```python
## saving the unique values from the exploded column
cols_to_make = exploded['Powers_split'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    df2[col] = df2['Powers'].str.contains(col)
df2.head()
```

```python
# drop Powers columns
df2 = df2.drop(columns=['Powers','Powers_split'])
```

```python
df2.head()
```

```python
df1.head()
```

```python
df1 = df1.iloc[:,[6,7,0,1,2,3,4,5,8,9]]
```

```python
df1.head()
```

```python
df2.head()
```

```python
## Combine into one final Dataframe
df = pd.merge(df1, df2, left_on='Hero', right_on='hero_names')
df.head()
```

We compare the average weight of superheros who have Super Speed to those who do not.

```python
df.groupby('Super Speed')['Weight'].mean()
```

On average those with super speed weigh almost 28 kg more than those without.
